```python
def splitData8020(X, Y):
    
    from sklearn.model_selection import train_test_split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state =33)
        
#     splitPoint = (int) (len(Y) * 0.8)
    
#     X_train = (X.iloc[:splitPoint, :])
#     Y_train = (Y.iloc[:splitPoint])
#     X_test = (X.iloc[splitPoint:, :])
#     Y_test = (Y.iloc[splitPoint:])
    
# #     print(X_train.head())
# #     print(Y_train.head())
#     print(X_test)
    
    X_train = X_train.values.tolist()
    Y_train = Y_train.values.tolist()
    X_test = X_test.values.tolist()
    Y_test = Y_test.values.tolist()

              
    return X_train, X_test, Y_train, Y_test
```

```python
def accuracyScore(predictions, Y_test):
    
    count = 0.0
    correct = 0.0
    for i, j in zip(predictions, Y_test):
        count += 1.0
        if i == j:
            correct += 1.0
    return correct / count
```

```python
os.getcwd()
zoo = pd.read_csv("/Users/Rong/Documents/USF/Machine Learning 2/zoo.csv", names = ['animal_name', 'hair', 'feathers', 'eggs', 'milk', 'airborne'
, 'aquatic', 'predator', 'toothed', 'backbone', 'breathes', 'venomous', 'fins', 'legs (numeric)', 'tail', 'domestic', 'catsize', 'type'])

X = zoo.iloc[:, 1:-1]
Y = zoo.iloc[:, -1]

X_train, X_test, Y_train, Y_test = splitData8020(X, Y)

# print(X_train)
# print(Y_train.shape())

predictions = []
my_model = decision_tree('entropy')
my_model.fit(X_train, Y_train)
predictions = my_model.predict(X_test)    
accuracy = accuracyScore(predictions, Y_test)

print("Accuracy score = " + str(accuracy))
```

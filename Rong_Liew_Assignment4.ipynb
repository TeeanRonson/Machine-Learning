```python
class knn_basic:
    
    #Constructor
    def __init__(self, K=3):
        self.K = K
    
        
    def fit(self, X, Y): 
        self.X = X
        self.Y = Y
        return 
    
    def predict(self, test):

        distances = self.findDistances(test);
        neighbours = self.findNeighbours(distances);
#         from collections import Counter
#         vote = Counter(neighbours).most_common(1)[0][0]

        vote = self.calcVote(neighbours)
        
        return vote
    
    def findDistances(self, test):
        
        distances = []
        for i in range(len(self.X)):
            euclidean_distance = np.sqrt(np.sum(np.square(test - self.X[i, :])))
            distances.append([euclidean_distance, i])
            
        distancesSorted = sorted(distances) 
#         print(distancesSorted)
        
        return distancesSorted
    
    def findNeighbours(self, distances):
        
        neighbours = []
        for i in range(self.K):
            index = distances[i][1]
#             print(self.Y[index])
            neighbours.append(self.Y[index])
        
        print(neighbours)
        return neighbours
    
    def calcVote(self, neighbours):
        votes = {}
#         print(len(neighbours))
        for i in range(len(neighbours)):
            group = neighbours[i]
            if group in votes:
                votes[group] +=1
            else:
                votes[group] = 1
#         print(vote)
        topVote = sorted(votes, key=votes.__getitem__, reverse= True)        
#         print(topVote)
        return topVote[0]
```

```python
def nearestNeighbours(X_train, Y_train, X_test, k):
    #fitting the model everytime I change the k value
    predictions = []
    new_model = knn_basic(k)
    new_model.fit(X_train, Y_train)

    for i in range(len(X_test)):
        predictions.append(new_model.predict(X_test[i, :]))

    return predictions
```

```python
def accuracyScore(predictions, Y_test):
    
    count = 0.0
    correct = 0.0
    for i, j in zip(predictions, Y_test):
        count += 1.0
        if i == j:
            correct += 1.0
    return correct / count
```
